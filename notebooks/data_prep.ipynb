```python
data = pd.read_csv('../data/form_answers.tsv', sep='\t')
data.head()
```

```python
columns_renaming_dict = {
    'Horodateur': 'Timestamp',
    'Dans quel Bachelor/Master êtes-vous inscrits? // What are you enrolled in at EPFL?  ': 'Enrollment',
    'En quelle année d’études êtes-vous actuellement ? // Which year are you in ?' : 'Year',
    'Pensez-vous que l’IA devrait être : // Do you think that AI should be :': 'Q1_AI_focus',
    'À quel point trouvez-vous pertinent de créer un bachelor spécifiquement dédié à l’IA ?// How relevant do you think it is to create a bachelor specifically tailored for AI?' : 'Q2_AI_bachelor_relevance',
    "Pensez vous que le bachelor SysCom est d'actualité et ne nécessite aucune réforme? // Do you think the current SysCom bachelor is up to date and needs no significant reform?" : 'Q3_SysCom_relevance',
    "Pensez-vous que les bachelors IN et SC devrait être fusionnés en un seul bachelor avec des \"tracks\" de blocs obligatoires, inspirés des spécialisations du nouveau master Informatique (Cyber / Systems / Foundations of computing / AI & Data Science) ?  Do you think that the IN and SC bachelors should be merged into a single bachelor with mandatory block \"tracks\", inspired by the specializations of the new Computer Science master\'s degree (Cyber ​​/ Systems / Foundations of computing / AI & Data Science)? " : 'Q4_IN_SC_merge',
    'Quel nom préférez-vous pour un Bachelor in AI ?  // What name do you prefer for a Bachelor in AI ? ' : 'Q5_Bachelor_name',
    'Pourriez-vous donner votre avis sur les changements apportés au nouveau Bachelor en IA ? (voir description ci-dessus) //  Can you please give your opinion on the changes for the new Bachelor in AI ! (see description above) [2nd Year changes]' : 'Q6_2nd_year_changes',
    'Pourriez-vous donner votre avis sur les changements apportés au nouveau Bachelor en IA ? (voir description ci-dessus) //  Can you please give your opinion on the changes for the new Bachelor in AI ! (see description above) [3rd Year changes -- MLII, ML Project, Intro to AI, AI application courses]' : 'Q7_3rd_year_changes_ML',
    'Pourriez-vous donner votre avis sur les changements apportés au nouveau Bachelor en IA ? (voir description ci-dessus) //  Can you please give your opinion on the changes for the new Bachelor in AI ! (see description above) [3rd Year changes -- making CompSec, StochMods, and PoDC optional]' : 'Q8_3rd_year_changes_removals',
    'Si vous aviez le choix entre le nouveau bachelor en IA et le bachelor en SysCom actuel, lequel des deux prendriez-vous? // If you had the choice to take either the current SysCom bachelor or the new reformed Bachelor in AI, which one would, you pick? ' : 'Q9_Bachelor_choice',
    'Commentaires sur changements // Comments on changes' : 'Comments'
    }
data_renamed = data.rename(columns=columns_renaming_dict)
data_renamed.head()
```

```python
cols = data_renamed.columns.tolist()[1:12] # select only enrollment and Q1 to Q9
distinct_answers = {col: data_renamed[col].unique() for col in cols}
```

```python
print("answers_renaming_map = {")
for col, answers in distinct_answers.items():
    print(f"\t\"{col}\": "+"{")
    for ans in answers:
        print(f"\t\t\"{ans}\": ,")
    print("\t},")
print("}")
```

```python
def q5_filter(str):
    """ Filters the Q5 answers into predefined categories. """
    try:
        lowered = str.lower()
    except AttributeError:
        return str
    if "data" in lowered:
        return "Data Science"
    if "intelligent" in lowered or "intelligence" in lowered:
        return "Intelligent Systems"
    if "ai" in lowered:
        return "AI"
    else:
        return "Other"

answers_renaming_map = {
    "Enrollment": {
		"Bachelor SysCom // Communication Systems": "BA-SC",
		"Master Cyber": "MA-CY",
		"Bachelor Informatique // Computer Science": "BA-IN",
		"Master Informatique // Computer Science": "MA-IN",
		"Master Data science": "MA-DS",
		"Other": "Other",
		"Master SysCom": "MA-SC",
	},
	"Year": {
		"BA3": "BA3",
		"MA1": "MA1",
		"BA5": "BA5",
		"MA3": "MA3",
		"Other (CMS/PhD/....)": "Other",
		"BA1": "BA1",
	},
	"Q1_AI_focus": {
		"Une spécialisation à l’intérieur d’un bachelor existant (ex. SysCom ou Informatique) //A specialization inside an existing bachelor": "Specialization",
		"Comme maintenant // As it is now": "As it is now",
		"Je ne sais pas / pas d’avis // I don't know / No opinion": "No opinion",
		"Un nouveau bachelor indépendant // A new independent bachelor": "New independent bachelor",
	},
	"Q2_AI_bachelor_relevance": {
		"Pertinent // Relevant": "Relevant",
		"Peu pertinent // A little Relevant": "A little Relevant",
		"Très pertinent // Very relevant": "Very relevant",
		"Pas d'opinion // No opinon": "No opinion",
		"Pas du tout pertinent // Not relevant at all": "Not relevant at all",
	},
	"Q3_SysCom_relevance": {
		"Pas d'opinion // No opinon": "No opinion",
		"D'accord // Agree": "Agree",
		"Pas d'accord // Disagree": "Disagree",
		"Pas du tout d'accord // Strongly disagree.": "Strongly disagree",
		"Tout à fait d'accord // Strongly agree": "Strongly agree",
	},
	"Q4_IN_SC_merge": {
		"Tout à fait d'accord // Strongly agree": "Strongly agree",
		"D'accord // Agree": "Agree",
		"Pas du tout d'accord // Strongly disagree.": "Strongly disagree",
		"Pas d'opinion // No opinon": "No opinion",
		"Pas d'accord // Disagree": "Disagree",
	},
	"Q5_Bachelor_name": q5_filter,  
	# "Q6_2nd_year_changes": {
	# 	"4": 4,
	# 	"1 – I hate it !": 1,
	# 	"5 – I love it !": 5,
	# 	"3 - No opinion": 3,
	# 	"nan": np.nan,
	# 	"2": 2,
	# },
	# "Q7_3rd_year_changes_ML": {
	# 	"5 – I love it !": 5,
	# 	"4": 4,
	# 	"3 - No opinion": 3,
	# 	"nan": np.nan,
	# 	"2": 2,
	# 	"1 – I hate it !": 1,
	# },
	# "Q8_3rd_year_changes_removals": {
	# 	"3 - No opinion": 3,
	# 	"1 – I hate it !": 1,
	# 	"4": 4,
	# 	"2": 2,
	# 	"nan": np.nan,
	# 	"5 – I love it !": 5,
	# },
	"Q9_Bachelor_choice": {
		"Nouveau Bachelor en IA // New AI Bachelor": "AI",
		"Bachelor actuel en SysCom // keep the current Communication Systems Bachelor": "SysCom",
		"Je ne suis pas sûr·e // I'm not sure": "Not sure",
	},
}
```

```python
data_cleaned = data_renamed.drop(columns=['Timestamp'])
for col, renaming_map in answers_renaming_map.items():
    data_cleaned[col] = data_cleaned[col].map(renaming_map)
data_cleaned.head()
```

```python
data_cleaned.to_csv('../data/form_answers_cleaned.tsv', sep='\t', index=False)
```
